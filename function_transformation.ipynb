<h3>square transform </h3>

```python
X = df.iloc[:,1:3]
y = df.iloc[:,0]
```

```python
trf = ColumnTransformer([('sq', FunctionTransformer(lambda x: x**2), ['Fare'])],remainder='passthrough')
```

```python
X_trans = trf.fit_transform(X)
```

```python
clf = LogisticRegression()

accuracy = np.mean(cross_val_score(clf,X_trans,y,scoring='accuracy',cv=10))

print(accuracy)
```
