```python
import numpy as np # for linear algebra
```

```python
import pandas as pd # Data processing
```

```python
import matplotlib.pyplot as plt 
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop('Purchased', axis = 1), df['Purchased'], test_size = 0.1, random_state = 0) 
```

```python
x_train.shape 
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train) # calculate the mean and standard deviation
```

```python
x_train_scaled = scaler.transform(x_train) # we transfrom both of the data sets test and and train 
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_ # prints the mean of the both cols (age and salary)
```

```python
x_train # its a pandas data frame 
```

```python
# adter standardization it got converted into numpy array
```

```python
x_train_scaled
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns = x_train.columns)
```

```python
x_test_scaled = pd.DataFrame(x_test_scaled, columns = x_test.columns)
```

```python
np.round(x_train.describe(), 1)
```

```python
np.round(x_train_scaled.describe(), 1)
```

<h2>Geometric Intuition</h2>

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Age'], x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Before Scaling')

sns.kdeplot(x_train['Age'], ax=ax1, label='Age')
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1, label='EstimatedSalary')

ax1.legend()


ax2.set_title('After Standard Scaling')

sns.kdeplot(x_train_scaled['Age'], ax=ax2, label='Age')
sns.kdeplot(x_train_scaled['EstimatedSalary'], ax=ax2, label='EstimatedSalary')

ax2.legend()

plt.show()
```

<h3>Comparision distribution</h3>

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)

ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(x_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```
